```python
import plotly.graph_objects as go
import numpy as np

N = 100000
r = np.random.uniform(0,1,N)
theta = np.random.uniform(0, 2*np.pi, N)

fig = go.Figure(data=go.Scattergl(
    x = r * np.cos(theta),
    y = r * np.sin(theta),
    mode='markers',
    marker=dict(
        color=np.random.randn(N),
        colorscale='Viridis',
        line_width=1
    )
))

fig.show()
```
